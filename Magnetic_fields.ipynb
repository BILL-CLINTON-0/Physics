### 1.2.2) Study of the streamline

For better visualiation we can replace the function `quiver` by the function `streamplot`, it will provide *field lines*, whose tangent are part of the *vector field* visualized with a quiver of red arrows in the figure above. In the case of magnetic fields, these are not called *force lines*.


```python
color = 2 * np.log(np.hypot(bx1f, by1f))

fig = plt.figure()
ax = fig.add_subplot(111)
ax.streamplot(X,Y,bx1f,by1f,color=color, linewidth=1, cmap=plt.cm.inferno,
              density=2, arrowstyle='->', arrowsize=1.5)
plt.title ('Mapping of Bx,By components un a xOy plane',color='b')


ax.add_artist(plt.Circle((0, 0), 0.001, color='g'))
ax.set_aspect('equal')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_xlim(-0.03,0.03)
ax.set_ylim(-0.03,0.03)
ax.set_aspect('equal')

plt.show()
```
